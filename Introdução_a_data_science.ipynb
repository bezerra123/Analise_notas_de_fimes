```python
import pandas as pd
import numpy as np

# Criando DataFrame de exemplo

dados = np.array([[100, 120, 90, 110, 130], [80, 150, 70, 140, 60]])

df = pd.DataFrame(dados, index = ["Semana 1", "Semana 2"],
                  columns= ["Segunda", "Tera", "Quarta", "Quinta", "Sexta"])

# Calculando o desvio padro das visitas por semana
desvio_padrao_semana1 = df.loc["Semana 1"].std()
desvio_padrao_semana2 = df.loc["Semana 2"].std()

print(df)
print("\nDesvio Padro da Semana 1:", desvio_padrao_semana1.round(2))
print("Desvio Padro da Semana 2:", desvio_padrao_semana2.round(2))
```
